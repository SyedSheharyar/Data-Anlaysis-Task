# Concat

       steps:
       1) filtering
       2) sort 
       3) head(3) 

```python
app_bool = laptops["Manufacturer"] == "Apple"
apple = laptops[app_bool]
```

```python
apple = apple.sort_values("Price Euros" , ascending = False).head(3)
```

```python
apple
```

```python
# For HP

app_bool = laptops["Manufacturer"] == "HP"
hp = laptops[app_bool]
hp = hp.sort_values("Price Euros" , ascending = False).head(3)
hp
```

```python
# concat

concat_df = pd.concat([apple , hp] , ignore_index = True)
concat_df
```

## Visualization

```python
import matplotlib.pyplot as plt
```

```python
# Data
Manufacturer = laptops["Manufacturer"].value_counts().index[:15]
counts = laptops["Manufacturer"].value_counts().values[:15]
percentage = round(laptops["Manufacturer"].value_counts(normalize = True)*100,1)[:15]


#Create stylish bar chart

plt.figure(figsize=(12,8))
bars = plt.bar(Manufacturer, counts, color="lightgrey",alpha=0.75, edgecolor="black", linewidth=1.5)

plt.xticks(rotation = -90, fontsize = 12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.grid(axis='x', linestyle="")
plt.xticks(fontsize=12) #Customized tick lables
plt.yticks(range(0,400,50),[],fontsize=12) #Customized tick lables and customized y-ticks range

max_count_manufacturer = Manufacturer[counts.argmax()]

max_count_index = list(Manufacturer).index(max_count_manufacturer)
bars[max_count_index].set_color('salmon')
bars[max_count_index].set_edgecolor('black')
bars[1].set_color('salmon')
bars[1].set_edgecolor('black')


for bar, perc in zip(bars,percentage):
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height + 2, "%d" % int(height), ha= "center", va="bottom", fontsize=10)
    plt.text(bar.get_x()+ bar.get_width()/2,height/2 ,f'{perc}%', ha= "center", va="bottom", fontsize=10, color="Blue")
    
ax=plt.gca()
ax.set_facecolor("#f7f7f7")

#Adding Chart title inside the chart
plt.text(0.5, 0.95, "Top Brands Customers Choice" , horizontalalignment= 'center', fontsize=16, transform=plt.gca().transAxes, color="grey", fontweight="bold")


#Adding Conclusion inside the chart
plt.text(0.5, 0.90, "The Dell And Lenovo Brands stand out at High Customer Choice" , horizontalalignment= 'center', fontsize=14, transform=plt.gca().transAxes, color="grey")

for i in ["top",'right','left']:
    plt.gca().spines[i].set_visible(False)

plt.tight_layout() #Adjust layout to prevant clipping

plt.show
```
